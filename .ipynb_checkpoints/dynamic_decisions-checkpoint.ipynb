```python
from keras.preprocessing import image as image_utils
from keras.applications.imagenet_utils import decode_predictions
from keras.applications.imagenet_utils import preprocess_input
from keras.applications import VGG16
from matplotlib import pyplot
from scipy.misc import imresize
import numpy
import argparse
import cv2
```

```python
def subimage(image, offset, dimension=(224,224)):
    x = int(offset[1] * image.shape[1])
    y = int(offset[0] * image.shape[0])
    sub = image[y:y + dimension[0], x:x + dimension[1], :].copy()
    sub = numpy.expand_dims(sub, axis=0)
    sub = preprocess_input(sub)
    return sub
```

```python
def plot_image(image):
    scaled_image = image.copy().squeeze()
    scaled_image -= scaled_image.min()
    scaled_image /= scaled_image.max()
    pyplot.imshow(scaled_image)
```

```python
image_filename = 'C:/Users/bimmy/Pictures/louis2.jpg'
offset = (0.5, 0.5)
```

```python
original = cv2.imread(image_filename)
image = image_utils.img_to_array(original)
```

```python
plot_image(preprocess_input(image))
```

```python
plot_image(subimage(image, offset))
```

```python
model = VGG16(weights="imagenet")
```

```python
predictions = model.predict(subimage(image, offset))
decoded = decode_predictions(predictions)
for i in range(5):
    print('{} {:.0%}'.format(decoded[0][i][1], decoded[0][i][2]))
```

```python
pyplot.plot(predictions.transpose())
pyplot.show()
```

```python
offsets = 0.1 * numpy.random.randn(100, 2) + 0.5
images = numpy.zeros((100, 224, 224, 3))
preds = numpy.zeros((100, 1000))
for t in range(100):
    images[t,:,:,:] = subimage(image, offsets[t])
preds = model.predict(images)
```

```python
pyplot.imshow(preds.transpose(), aspect='auto')
mean_preds = preds.max(axis=0).reshape((1, 1000))
order = numpy.argsort(mean_preds)
decoded = decode_predictions(mean_preds)
```

```python
for i in range(5):
    pyplot.plot(preds[:,order[0][-1 - i]].transpose(), label=decoded[0][i][1])
pyplot.legend()
pyplot.show()
```
